```python
df = pd.read_csv('../mat305_2024.csv')
df.columns = ['prénom', 'nom', 'CC1', 'CC2', 'moyenne', 'note TD']
df
```

```python
xdf = df.dropna()
```

```python
sum(xdf['note TD'].values)/len(xdf)
```

```python
a, b = xdf['CC2'].values, xdf['note TD'].values
vv = stats.linregress(a,b)

ab = np.array([5,20])

plt.plot(a,b,'r.')
plt.plot(ab, vv.slope*ab + vv.intercept);
```
